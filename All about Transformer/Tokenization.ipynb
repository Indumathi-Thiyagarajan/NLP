```python
import tiktoken
```

```python
from transformers import AutoModel, AutoTokenizer

def show_tokens(sentence, tokenizer_name, print_token_ids = False):
  if tokenizer_name == 'gpt-4':
    tokenizer = tiktoken.encoding_for_model("gpt-4")
    token_ids = tokenizer.encode(sentence)

  else:
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids

  for t in token_ids:
    if print_token_ids:
      print(t, end= " ")
    print('\x1b[0;30;47m' + tokenizer.decode([t]) + '\x1b[0m', end = ' ')

  print('\n\n')

  for t in token_ids:
    print(t,'\x1b[0;30;47m' + tokenizer.decode([t]) + '\x1b[0m')

def encode_decode(sentence, tokenizer_name):
  tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
  token_ids = tokenizer(sentence).input_ids
  print(tokenizer.decode(token_ids))


text = """
English and CAPITALIZATION
😀😀
show tokens False None elif == >= else:
Two tabs:"\t\t" Four spaces: "  "
12.0 *50 = 600
"""
```

cannot new line,  see emoji and  see capitalization. So we should not use this model that considers above items as important information not be lost

```python
show_tokens(text, 'bert-base-uncased')
```

```python
encode_decode(text, 'bert-base-uncased')
```

Capitalization is preserved, cannot see emoji and new line

```python
show_tokens(text, 'bert-base-cased')
```

new line is preserved , emoji is preserved and capitalization is preserved
